```python
#Get Polling Unit summary
import pandas as pd
import numpy as np

latest_voter_summary_dir = '/Users/carbz/Dropbox/Nigeria-NEAP-GIS/shapefiling/voter_sum_of-voter_04pu_summary-concat-FINAL.csv'
pu_summary = pd.read_csv(latest_voter_summary_dir)

pu_summary.columns = ['State',
                        'LGA',
                        'Ward',
                        'PollingUnit',
                        'BoothId',
                        'registered_voters',
                        'source_file']

#print(pu_summary.shape) #dimensions of dataset
pu_summary
```

```python
pieces_of_interest = pu_locations.reindex(columns=['BoothId','BoothLat','BoothLong'])
pieces_of_interest_full = pu_locations.reindex(columns=['BoothId','BoothLat','BoothLong',
                                                        'state','lga','ward','Booth','BoothDescription'])

pieces_of_interest_full.columns = ['BoothId','BoothLat',
                                   'BoothLong','state-NEC', 
                                   'lga-NEC','ward-NEC',
                                   'Booth-NEC','BoothDescription']

pieces_of_interest_full
```
