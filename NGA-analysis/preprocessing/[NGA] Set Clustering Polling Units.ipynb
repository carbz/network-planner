```python

# Function for the MST & distance matrix
#build a distance martrix (edge weights)

import numpy as np
from numpy import sin, cos, pi, arcsin, sqrt

def hav_dist(vector,point):
    lat = vector[:, 0]
    lon = vector[:, 1]
    x, y = point
    return get_hav_distance(lat, lon, x, y)

def get_hav_distance(lat, lon, pcode_lat, pcode_lon):
    """
    Find the distance between a vector of (lat,lon) and the reference point (pcode_lat,pcode_lon).
    """
    rad_factor = pi / 180.0  # degrees to radians for trig functions
    lat_in_rad = lat * rad_factor
    lon_in_rad = lon * rad_factor
    pcode_lat_in_rad = pcode_lat * rad_factor
    pcode_lon_in_rad = pcode_lon * rad_factor
    delta_lon = lon_in_rad - pcode_lon_in_rad
    delta_lat = lat_in_rad - pcode_lat_in_rad
    # Next two lines is the Haversine formula
    inverse_angle = (np.sin(delta_lat / 2) ** 2 + np.cos(pcode_lat_in_rad) *
                     np.cos(lat_in_rad) * np.sin(delta_lon / 2) ** 2)
    haversine_angle = 2 * np.arcsin(np.sqrt(inverse_angle))
    earth_radius =  6371010 # meters
    return haversine_angle * earth_radius


# """Returns the computed distance matrix"""
def dist_mat(coords):
    """
    returns the distance matrix of a numpy array (haversine)
    """
    return np.row_stack(map(lambda pt: hav_dist(coords, pt), coords))
    
def connected_graph_mst(coords, plot=True):
    dmat = dist_mat(coords)
    G = nx.from_numpy_matrix(dmat)
    nx.set_node_attributes(G, 'coords', {i:coord for i, coord in enumerate(coords)})
    mst = nx.algorithms.minimum_spanning_tree(G)
    
    if plot:
        import matplotlib.pyplot as plt
        fig = plt.figure(figsize=(10,10))
        ax = fig.add_subplot(111)
        nx.draw(mst, nx.get_node_attributes(mst, 'coords'), node_size=40, ax=ax)
    return mst
```

```python
from statistics import mean

def LV_MSTs(dir_of_setcovered_hhs, out_dir):
    '''
    Determine and output LV networks as aggregated about centroids
    
    Input: 
    (i)a Directory location of a csv file.  File expected to have nodes pre-clustered
    Expected values include a cluster_id, central_node, x, y, node_id and weight_households
    Ex: r'/Users/carbz/Dropbox/Nigeria-NEAP-GIS/Cluster_vs_PollingUnits_Analysis/Buf2/clustering-set_cover/All_Points-SetCovering-search_radius_1000m.csv'
    (ii) A specification for an output directory ex.
    Ex: r'/Users/carbz/Dropbox/Nigeria-NEAP-GIS/Cluster_vs_PollingUnits_Analysis/Buf2/clustering-set_cover/'        

    Output: a csv in same directory with aggregations of clusters about a centroid
    '''
    
    #First import our csv
    print('importing clustered csv')
    df = pd.read_csv(dir_of_setcovered_hhs, header = 0)
    
    #Defining output column names and blank dataframe
    columns = ['cluster_id','mst_length_m', 'compounds_qty','x_centroid','y_centroid','mst_length_per_compound_m']
    mst_df = pd.DataFrame(columns=columns)
    
    G1 = nx.Graph() #empty networkx graph 
    total_LV = 0 #define a running LV count
    
    #Unique clusters to aggregate
    clusters=(df.cluster_id.unique())

    for i in clusters:
        
        #Do not consider the 'noise' clusters
        if i >=  0:

            pts = (df[df['cluster_id'] == i][['x','y']].values)
            
            mst_i = connected_graph_mst(pts, plot=False)

            mst_edges = mst_i.edges()

            #Manually calculate the MST length of each cluster
            mst_length = 0 #reset at start 

            if len(mst_edges) >0:
                for (u, v) in mst_edges:
                    x1=pts[u][0]
                    y1=pts[u][1]
                    x2=pts[v][0]
                    y2=pts[v][1]

                    d = get_hav_distance(x1, y1,x2,y2)

                    mst_length=d + mst_length

            compounds_qty = len(pts)

            x_centroid = (df[df['cluster_id'] == i][['x']].values).mean()
            y_centroid = (df[df['cluster_id'] == i][['y']].values).mean()

            summary = pd.Series({'cluster_id':i,
                                 'mst_length_m':mst_length,
                                 'x_centroid': x_centroid,
                                 'y_centroid': y_centroid,
                                 'compounds_qty':compounds_qty,
                                 'mst_length_per_compound_m':(mst_length/compounds_qty)})

            mst_df.loc[i] = summary

            G2 = mst_i
            G1=nx.disjoint_union(G1, G2)
            
            #Update the running length of LV needed
            total_LV = total_LV + mst_length

#     file_name = '/Users/carbz/Dropbox/Nigeria-NEAP-GIS/Cluster_vs_PollingUnits_Analysis/Buf1/clustering-set_cover/Clustered_HHs-500m_SetCover-20150203.csv'      
    mst_df.to_csv(out_dir, index = False)
    
    print('MST length total (in km) is: '+str(total_LV/1000))
```

```python
grid_pts_dir = '/Users/carbz/Dropbox/Nigeria-NEAP-GIS/Cluster_vs_PollingUnits_Analysis/Buf1/clustering-set_cover/Clustered_HHs-500m_SetCover-20150203.csv'  

def csv_to_mst(grid_pts_dir):
    '''
    Function takes as an input a directory of a csv containing 'x','y' coordinated
    example file
    # grid_pts_dir = '/Users/carbz/Dropbox/Nigeria-NEAP-GIS/Cluster_vs_PollingUnits_Analysis/Buf1/clustering-set_cover/Clustered_HHs-500m_SetCover-20150203.csv'  
    outputs the sum of a minimum spanning tree to connect all points
    '''
    #read dataframe in 
    df = pd.read_csv(grid_pts_dir, header = 0)
    
    #ensure x, y exist and trim out other data
    if ('x_centroid' in df.columns):
        df['x'] = df['x_centroid']
        df['y'] = df['y_centroid']
    #adjust for caps
    if ('X' in df.columns):
        df['x'] = df['X']
        df['y'] = df['Y']
    pts = (df[['x','y']].values)
    
    #utilize networkx to calculate MST
    mst_length=0 #at start
    mst = connected_graph_mst(pts) 
    mst_edges = mst.edges()

    #calculate & tally up length of each edge
    if len(mst_edges) >0:
        for (u, v) in mst_edges:
            x1=pts[u][0]
            y1=pts[u][1]
            x2=pts[v][0]
            y2=pts[v][1]

            d = get_hav_distance(x1, y1,x2,y2)

            mst_length=d + mst_length
    
    return(mst_length/1000, len(df))
```
